# Tripple Coin Flip
In the following is a quantum circuit with three qubits that can act as a tripple coin flip. 

## 1. Import necessary libraries

```python
from qiskit import *
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile
from qiskit.visualization import plot_histogram
```

## 2. Create the circuit
2.1 crate a circuit with three qubit and three classical bit

2.2 Apply Hadamard gate to the all three qubits (superposition)

2.3 Measure each qubit and write to corresponding classic bit

(Optional: draw the circuit)

```python
qc = QuantumCircuit(3, 3)

qc.h(0)
qc.h(1)
qc.h(2)

qc.measure(0, 0)
qc.measure(1, 1)
qc.measure(2, 2)

qc.draw('mpl')
```

## 3. Simulate the circuit
3.1 Select the simulator

3.2 Transpile the quantum circuit in elementary gates

3.3 Run the circuit

3.4 Show the results

```python
simulator = AerSimulator()

compiled_circuit = transpile(qc, simulator)

result = simulator.run(compiled_circuit, shots=1000).result()

counts = result.get_counts()
print("Results:", counts)
plot_histogram(counts, title="Coin Flip Distribution")
```
